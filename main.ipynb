```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load images from the specified directory into a TensorFlow dataset.
# The function automatically labels images based on their subdirectory names.
# Here, "data_dir" is the path to the directory containing the image dataset.
data = tf.keras.utils.image_dataset_from_directory(data_dir) #building our data pipeline
```

```python
# Convert the dataset to a NumPy iterator for easier manipulation
data_iterator = data.as_numpy_iterator() # accessing our data pipeline

# This is important because it allows us to access the dataset in a more flexible way.
# Using a NumPy iterator, we can directly retrieve batches of images and labels,
# enabling us to inspect, preprocess, or visualize the data as needed.
# This is especially useful for debugging or understanding the structure of the dataset.
```

```python
batch = data_iterator.next()  # Get the next batch of images and labels
# The "next()" method retrieves the next batch of data from the iterator.
# This batch contains both images and their corresponding labels.
# The batch is a tuple containing two elements: images and labels.
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20, 20))
for idx, img in enumerate(batch[0][:4]):  # Iterate through the images in the batch
    ax[idx].imshow(img.astype(int))  # Display the image
    ax[idx].set_title(batch[1][idx])  # Set the title to the corresponding label
```

```python
data = data.map(lambda x, y: (x / 255, y))  # Normalize the images to the range [0, 1]
# Normalizing the images is a common preprocessing step in deep learning.
# It helps to improve the convergence of the model during training.
# By scaling the pixel values to the range [0, 1], we ensure that the model
```

```python
scaled_iterator = data.as_numpy_iterator()
# receives input values that are more manageable and consistent.
# This can lead to faster training and better performance.
```

```python
batch = scaled_iterator.next()  # Get the next batch of images and labels
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20, 20))
for idx, img in enumerate(batch[0][:4]):  # Iterate through the images in the batch
    ax[idx].imshow(img)  # Display the image
    ax[idx].set_title(batch[1][idx])  # Set the title to the corresponding label
```

```python
len(data)
```

```python
train_data = int(len(data)*.7)  # Take the first 80% of the dataset for training
val_data = int(len(data)*.2)  # Skip the first 80% for validation
test_data = int(len(data)*.1) + 1  # Skip the first 90% for testing
train_data, val_data, test_data
```

```python
train = data.take(train_data)  # Take the first 80% of the dataset for training
val = data.skip(train_data).take(val_data)  # Skip the first 80% for validation
test = data.skip(train_data + val_data).take(test_data)  # Skip the first 90% for testing
len(train), len(val), len(test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
```

```python
model = Sequential([
    Conv2D(16, (3, 3), 1, activation='relu', input_shape=(256, 256, 3)),  # First convolutional layer
    MaxPooling2D(pool_size=(2, 2)),  # First max pooling layer
    Conv2D(32, (3, 3), 1, activation='relu'),  # Second convolutional layer
    MaxPooling2D(pool_size=(2, 2)),  # Second max pooling layer
    Conv2D(16, (3, 3), 1, activation='relu'),  # Third convolutional layer
    MaxPooling2D(pool_size=(2, 2)),  # Third max pooling layer
    Flatten(),  # Flatten the output from the convolutional layers
    Dense(256, activation='relu'),  # Fully connected layer with ReLU activation
    Dropout(0.5),  # Dropout layer to reduce overfitting
    Dense(1, activation='sigmoid')  # Output layer with sigmoid activation for binary classification (0 0r 1)
])
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(optimizer=Adam(learning_rate=0.001),  # Adam optimizer with a learning rate of 0.001
              loss='binary_crossentropy',  # Binary cross-entropy loss for binary classification
              metrics=['accuracy'])  # Accuracy metric for evaluation
model.summary()  # Print the model summary to see the architecture and parameters
# The model is a convolutional neural network (CNN) designed for binary classification tasks.
```

```python
logdir = "/app/src/logs"
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)  # TensorBoard callback for logging
# This callback allows us to visualize the training process in TensorBoard.
```

```python
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])  # Train the model
```

```python
fig = plt.figure(figsize=(20, 10))
plt.plot(hist.history['accuracy'], label='accuracy', color = "teal")  # Plot training accuracy
plt.plot(hist.history['val_accuracy'], label='val_accuracy', color = "orange")  # Plot validation accuracy
fig.suptitle('accuracy', fontsize=20)
plt.legend(loc = "upper left")
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
plt.plot(hist.history['loss'], label='loss', color = "teal")  # Plot training accuracy
plt.plot(hist.history['val_loss'], label='val_loss', color = "orange")  # Plot validation accuracy
fig.suptitle('loss', fontsize=20)
plt.legend(loc = "upper right")
plt.show()
```

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
rec = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test:
    images, labels = batch
    predictions = model.predict(images)
    pre.update_state(labels, predictions)
    rec.update_state(labels, predictions)
    acc.update_state(labels, predictions)
print("Precision: ", pre.result().numpy())
print("Recall: ", rec.result().numpy())
print("Accuracy: ", acc.result().numpy())
# The model is evaluated on the test dataset to assess its performance.
# The precision, recall, and accuracy metrics are calculated and printed.
```
